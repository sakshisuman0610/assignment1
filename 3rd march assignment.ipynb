Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter plot: A scatter plot is used to visualize the relationship between two variables. It is used to identify the correlation between variables and to identify any outliers in the data.

Pairplot: It draws relationship between all numerical entries .

Catplot: Categorical plots are used to visualize the relationship between a categorical variable and a numerical variable.

Line plot: A line plot is used to visualize trends over time. It is used to display how a variable changes over time.

Histogram: A histogram is used to visualize the distribution of a single variable. It is used to identify the frequency of occurrence of values in a dataset.

Heatmap: A heatmap is used to visualize the correlation between variables. It is used to identify the strength and direction of the relationship between variables.

Box plot: A box plot is used to visualize the distribution of a dataset. It is used to identify outliers and to compare the distribution of multiple datasets.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'. Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')


sns.boxplot(x='pclass', y='age', data=titanic)
sns.boxplot(x='pclass', y='fare', data=titanic)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')

sns.histplot(data=diamonds,x="price",hue="cut")
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris=sns.load_dataset("iris")
sns.pairplot(data=iris,hue="species")
```
